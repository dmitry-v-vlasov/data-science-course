```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import model_selection
from sklearn import linear_model
from sklearn import tree
from sklearn import ensemble
from sklearn import metrics
```

```python
data_weather = pd.read_csv('data/weatherAUS.csv')
data_weather.head(10)
```

```python
pd.set_option("display.max_columns", None)
# data_weather.drop(['Date'], axis=1).sample(10)
```

```python
data = data_weather.drop(['Evaporation', 'Sunshine', 'Cloud3pm'], axis=1)
data.head(10)
```

```python
yes_no_binary = lambda value: value if not isinstance(value, str) else 1 if value.lower() == 'yes' else 0 if value.lower() == 'no' else value

data['RainToday'] = data['RainToday'].apply(yes_no_binary)
data['RainTomorrow'] = data['RainTomorrow'].apply(yes_no_binary)

data['Date'] = pd.to_datetime(data['Date'])
data['month'] = data['Date'].dt.month
data.drop(['Date'], axis=1, inplace=True)

display(data['RainToday'].mean())
display(data.head(10))
```

```python
data_season = data.groupby('month').mean()
data_season[['RainToday']]
```

```python
data.info()
```

```python
categorical = ['month', 'Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm']
dataset = pd.get_dummies(data, columns=categorical)
dataset.columns
```

```python
dataset.columns[dataset.columns.str.contains('month')]
```

```python
dataset.head()
```

```python
dataset_ready = dataset.copy(deep=True)

dataset_ready.dropna(axis=0, inplace=True)
dataset_ready.reset_index(drop=True, inplace=True)
dataset_ready.head()
```

```python
dataset_ready.shape
```

```python
X = dataset_ready.drop(['RainTomorrow'], axis=1)
y = dataset_ready['RainTomorrow']

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=31)
X_train.reset_index(drop=True, inplace=True)
X_test.reset_index(drop=True, inplace=True)
y_train.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)
display(np.mean(y_test))
```

```python
np.random.seed(31)
# np.random.randint()
train_index_min = X_train.index.min()
train_index_max = X_train.index.max()
N_train = len(X_train)
display(N_train)

bootstrap_sets = list()
mean_min_temps = list()
N_bootstraps = 1000
for ixb in range(1, N_bootstraps + 1):
    bootstrap_index = [np.random.randint(train_index_min, train_index_max) for it in range(1, N_train + 1)]
    bootstrap_set = X_train.iloc[bootstrap_index]
    # display(bootstrap_set)
    bootstrap_sets.append(bootstrap_set)
    mean_min_temps.append(bootstrap_set['MinTemp'].mean())

np.std(mean_min_temps)
```

```python
model = linear_model.LogisticRegression(random_state=42)
model.fit(X_train, y_train)

y_test_pred = model.predict(X_test)

metrics.roc_auc_score(y_test, y_test_pred)
```

```python
params = {
    'max_leaf_nodes': list(range(2, 10)),
    'min_samples_split': [2, 3, 4],
    'max_depth': [5,7,9,11]
}

basic_model = tree.DecisionTreeClassifier(random_state=42)
searcher = model_selection.GridSearchCV(estimator=basic_model, param_grid=params, scoring='roc_auc', cv=3, n_jobs=4)
searcher.fit(X_train, y_train)

display(searcher.best_params_)
display(searcher.best_estimator_)

y_test_pred = searcher.best_estimator_.predict(X_test)
display(metrics.roc_auc_score(y_test, y_test_pred))
```

```python
params = {'max_leaf_nodes': list(range(5, 15)), 'min_samples_split': [2, 3, 4], 'max_depth': [5,7,9,11]}
grid_search_cv = model_selection.GridSearchCV(tree.DecisionTreeClassifier(random_state=42), params, cv=3, n_jobs=4)
grid_search_cv.fit(X_train, y_train)
grid_search_cv.best_params_
clf = tree.DecisionTreeClassifier(max_depth = 5, max_leaf_nodes = 9, min_samples_split = 2, random_state=42)
clf.fit(X_train, y_train)
preds_train = clf.predict(X_train)
preds_test = clf.predict(X_test)
metrics.roc_auc_score(preds_test, y_test)
```

```python
model = ensemble.RandomForestClassifier(n_estimators=100, random_state=31, n_jobs=4)
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)
metrics.roc_auc_score(y_test, y_test_pred)
```

```python
params = {'max_features': [ 4, 5, 6, 7], 'min_samples_leaf': [3, 5, 7, 9, 11], 'max_depth': [5, 10, 15], 'n_estimators': [100]}
basic_model = ensemble.RandomForestClassifier(random_state=31, n_jobs=4)

searcher = model_selection.GridSearchCV(estimator=basic_model, param_grid=params, scoring='roc_auc', cv=3, n_jobs=4)
searcher.fit(X_train, y_train)

display(searcher.best_params_)
display(searcher.best_estimator_)
```

```python
clf = ensemble.RandomForestClassifier(n_estimators=100, max_depth=15, max_features=7, min_samples_leaf=3, n_jobs=4, random_state=31)
clf.fit(X_train, y_train)
y_test_pred = searcher.best_estimator_.predict(X_test)
metrics.roc_auc_score(y_test, y_test_pred)
```
