```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import preprocessing
from sklearn import model_selection
from sklearn import linear_model
from sklearn import tree
from sklearn import neighbors
from sklearn import naive_bayes
from sklearn import ensemble
from sklearn import metrics
```

```python
pd.set_option('display.max_columns', None)
```

```python
data_frogs = pd.read_csv('data/Frogs_MFCCs.csv')
data_frogs.head()
```

```python
data = data_frogs.drop(['Genus', 'Species', 'RecordID'], axis=1)
data.head()
```

```python
data['Family'].value_counts()
```

```python
data['Family'] = data['Family'].apply(lambda v: 1 if v == 'Dendrobatidae' else 0)
data['Family'].value_counts()
```

```python
X = data.drop(['Family'], axis=1)
y = data['Family']

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=31)
display(X_train.shape)
display(X_test.shape)
```

```python
model = ensemble.RandomForestClassifier(n_estimators=10, random_state=42)
model.fit(X_train, y_train)
y_test_pred = model.predict(X_test)
metrics.f1_score(y_test, y_test_pred)
```

```python
models = [
    ('rf', ensemble.RandomForestClassifier(n_estimators=10, random_state=31)),
    ('KNN', neighbors.KNeighborsClassifier(n_neighbors=11)),
    ('nb', naive_bayes.GaussianNB())
]

stacking_model = ensemble.StackingClassifier(estimators=models, final_estimator=linear_model.LogisticRegression())
stacking_model.fit(X_train, y_train)
y_test_pred = stacking_model.predict(X_test)
metrics.f1_score(y_test, y_test_pred)
```
